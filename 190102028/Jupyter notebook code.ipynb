```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt 
```

```python
url = "https://raw.githubusercontent.com/eshantrehan26/Learning-Content/master/Phase%203%20-%202020%20(Summer)/Week%201%20(Mar%2028%20-%20Apr%204)/assignment/data.txt"
df = pd.read_table(url)
df
```

```python
df_label1 = df[df.Label == 1]
df_label2 = df[df.Label == 2]
```

```python
for i1 in range(1, 11, 1):
    for i2 in range(i1 + 1, 11, 1):
        plt.scatter(df_label1[str(i1)], df_label1[str(i2)],c = 'red')
        plt.scatter(df_label2[str(i1)], df_label2[str(i2)],c = 'blue')
        plt.xlabel("Feature " + str(i1))
        plt.ylabel("Feature " + str(i2))
        plt.title(str(i2)+" vs " + str(i1))
        plt.savefig(str(i2)+" vs " + str(i1) +".png")
        plt.show()
```

```python
from sklearn.preprocessing import StandardScaler
features = [str(i1) for i1 in range (1, 11, 1)]
y = df.loc[:,'Label'].values
X = df.loc[:,features].values
X = StandardScaler().fit_transform(X)
df_X = pd.DataFrame(X, columns = features)
```

```python
df_X
```

```python
from sklearn.decomposition import PCA
result = []
for i1 in range(1, 11, 1):
    for i2 in range(i1 + 1, 11, 1):
        pca = PCA(n_components= 1)
        values = df_X.loc[:,[str(i1),str(i2)]].values
        principle_values = pca.fit_transform(values)
        temp = pca.explained_variance_ratio_
        result.append([str(i1) + " vs " + str(i2), temp[0]])
result_df = pd.DataFrame(result)
```

```python
principle_df = pd.DataFrame(principle_values)
```
