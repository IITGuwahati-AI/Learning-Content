```python
import matplotlib.pyplot as plt
```

```python
import numpy as np
```

```python
v=np.zeros((11,1000))
```

```python
v=np.loadtxt('data.txt')
```

```python
a=np.arange(1,11)
```

```python
b=1
c=2
x=np.zeros((1000,1))
y=np.zeros((1000,1))
p=np.zeros((1000,1))
q=np.zeros((1000,1))
```

```python
for b in a:
    c=b+1
    while(c<11):
        j=0
        k=0
        for num in a:
            
            if v[num,0]==1:
                x[j]=v[num,b]
                y[j]=v[num,c]
                j=j+1
                
            if v[num,0]==2:
                p[k]=v[num,b]
                q[k]=v[num,c]
                k=k+1
                
                
        plt.title('LABEL')
        plt.plot(x,y,'r',label='label 1')
        plt.plot(p,q,'c',label='label 2')
        plt.ylabel('feature')
        plt.xlabel('feature')
        plt.legend()
        plt.show()
        c=c+1
```
