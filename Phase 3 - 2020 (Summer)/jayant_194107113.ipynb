```
#Import the dataset into a matrix using numpy. The dataset is space delimited with first column as labels and rest 10 as features.

import numpy as np
data=np.genfromtxt("/content/data.txt",skip_header=1)
labels=data[:,0]
features=data[:,10]
feature_1=data[:,1]

import pandas as pd
df=pd.read_csv("/content/data.txt",delimiter='\t',skiprows=0)
df_sort=df.sort_values(by='Label',axis=0)
df_label1=df[df['Label']==1]
df_label2=df[df['Label']==2]
df_label1=df_label1.values
df_label2=df_label2.values
```

```
#Plot all features vs each other (2 at a time) using matplotlib - red for label 1 and blue for label 2, 
#clearly denote the axes and the title of the plot along with proper markings (what does each axis represent) on the axes.

from matplotlib import pyplot as plt
for i in range(1,11):
  for j in range(1,11):
    if (i!=j):
      plt.scatter(df_label1[:,i],df_label1[:,j],c='r')
      plt.scatter(df_label2[:,i],df_label2[:,j],c='b')
      #plt.xlable(f"feature_{i}")
      #plt.ylable(f"feature_{j}")
      plt.title(f"feature_{i} vs feature_{j}")
      plt.show()
      plt.xlabel(f"feature_{i}")
      plt.ylabel(f"feature_{j}")
```
