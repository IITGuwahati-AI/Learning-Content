The first plot between Feature 3 and Feature 8 classifies the two labels perfectly

```python
import numpy as np
from matplotlib import pyplot as plt

data = np.loadtxt('data.txt', delimiter='\t', skiprows=1)
#count = 1

# for a in range(1, 11):
#     for b in range(a+1, 11):

dict1 = {}
dict2 = {}

for k in range(0, data.shape[0]):
    if data[k][0] == np.float64(1):
        dict1[data[k][3]] = data[k][8]

    else:
        dict2[data[k][3]] = data[k][8]

x1 = []
y1 = []

for i in sorted(dict1):
    x1.append(i)
    y1.append(dict1[i])

x2 = []
y2 = []

for i in sorted(dict2):
    x2.append(i)
    y2.append(dict2[i])

plt.scatter(x1, y1, color = 'r', label = "Label 1")
plt.scatter(x2, y2, color = 'b', label = "Label 2")

plt.title('Feature {} v/s Feature {}'.format(8, 3))
plt.ylabel('Feature {}'.format(8))
plt.xlabel('Feature {}'.format(3))
plt.legend()
plt.show()
#print(count)
#count+=1
```
