# Question 4
We now incoroporate **PCA (Principal Component Analysis)** in the following dataset. We perform PCA on both standardized data as well as normal dataset and plot the two variants. We observe that the results of PCA and manual plotting differ. Better results are obtained only from $X_1$ and $X_2$ plot.\
Standardization or normalization of data is done to bring all the features to the same mean and variance (0,1):
$$Z = \frac{X-\mu}{\sigma}$$

**Answer:** We find that the plot for the normal data resembles the plot between third and tenth feature $X_3$ and $X_{10}$ with the axes swapped.

```
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA

## Create pca object
pca = PCA(n_components = 2)
## For normal data:
## Compute principal components and plot them
principalComponents = pca.fit_transform(data[:,1:11])
print("Principal components for Normal data: ",principalComponents)
principalComponent_1 = principalComponents[:,0]
principalComponent_2 = principalComponents[:,1]
print()
for n in range(999):
  if labels[n] == 2:
    plt.plot(principalComponent_1[n], principalComponent_2[n], 'bo')
  else:
    plt.plot(principalComponent_1[n], principalComponent_2[n], 'ro')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.title('Distribution of Normal data')
plt.show()
print()
## For standardized data:
## We standardize the data
## Compute principal components and plot them:
standardized_data = StandardScaler().fit_transform(data[:,1:11])
principalComponents = pca.fit_transform(standardized_data)
print("Principal components for Standardized data: ",principalComponents)
principalComponent_1 = principalComponents[:,0]
principalComponent_2 = principalComponents[:,1]
print()

for n in range(999):
  if labels[n] == 2:
    plt.plot(principalComponent_1[n], principalComponent_2[n], 'bo')
  else:
    plt.plot(principalComponent_1[n], principalComponent_2[n], 'ro')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.title('Distribution of Standardized data')
plt.show()


```
