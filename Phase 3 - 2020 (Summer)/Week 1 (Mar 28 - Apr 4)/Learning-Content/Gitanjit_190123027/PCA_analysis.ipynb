```python
import numpy as np
from matplotlib import pyplot as plt
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
data=np.loadtxt(fname='data_wk1')
x=data[:,0]
Y=data[:,1:11]
maxk=0
maxl=0
itr=[0,0]
itrl=[0,0]
from sklearn.preprocessing import StandardScaler
p = StandardScaler().fit_transform(data)
yp=p[:,1:11]
r=p[:,0:1]
for i in range(0,10):
    for j in range(i+1,10):
        x1=yp[:,i:i+1]
        x2=yp[:,j:j+1]
        
        f=np.concatenate((x1,x2),axis=1)
        pca=PCA(n_components=1)
        principalComponents1=pca.fit_transform(f)
        k=pca.explained_variance_ratio_
        g=np.concatenate((r,principalComponents1),axis=1)
        principalComponents1=pca.fit_transform(g)
        l=pca.explained_variance_ratio_
        
        if k>maxk:
            maxk=k
            itr=[i+1,j+1]
            
        if l>maxl:
            maxl=l
            itrl=[i+1,j+1]
        print(str(i+1)+' '+str(j+1)+' '+str(k)+' '+str(l))

print(maxk)
print(itr)
print(maxl)
print(itrl)
```

```python
print("From the above analysis it is concluded that features 3 and 8 retain the largest variance ratio when PCA-reduced from 2-d to 1-D")
print("That means that this feature pair is the most alike and hence redundant i.e. one feature can be expressed in the other's form  ")
print("But when combined with the labels: 1,2 retains the highest variance ratio which means that it ,this pair, can depict the labels in the best possible way ")
```

```python
#in order to look at the reduced data and understand which feature pairs depict the data in the best way , the following code segment:
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
data=np.loadtxt(fname='data_wk1')
X=data[:,0]
Y=data[:,1:11]
m = X.shape[0]#999
c= 0


for i in range (0,10):
    for j in range (i+1,10):
        x1 = Y[:,i:i+1]
        x2 = Y[:,j:j+1]
        
        x=np.concatenate((x1,x2),axis=1)
        xd = pd.DataFrame.from_records(x)
        xd = StandardScaler().fit_transform(x)
        pca = PCA(n_components = 1)
        principalComponent = pca.fit_transform(xd)
        st1 = 'Principal Component ' + str(c + 1)
        principalDf = pd.DataFrame(data = principalComponent, columns = [st1])
        prin = principalDf.to_numpy()
        for g in range (0,m):
            if X[g] == 1:
                plt.scatter(prin[g],X[g], color='red')
            else:
                plt.scatter(prin[g],X[g],color='blue')
        str6 = 'P.C. from feature ' + str(i+1) + ' and feature ' + str(j+1)
        plt.xlabel('One dimensional reduced array')
        plt.ylabel("Labels")
        plt.title(str6)
        
        plt.show()
```

```python
print("From the above analysis its clear from the figures that feature 1 and 2 pair distinguishes the labels most appropriately")
print("The second best pair seems to be 2,10 which matches with the first analysis made above [91.99 for feature 2,10 pair and 92.11 for 1,2 pair}")
```
