```python
import pandas as pd
```

```python
df = pd.read_csv('data.txt', delim_whitespace=True)
```

```python
#Renaming the columns 
df.columns = ['Label','a','b','c','d','e','f','g','h','i','j']
```

```python
df
```

```python
from sklearn.preprocessing import StandardScaler
count = 0
for var in 'abcdefghij':
    for letr in 'abcdefghij':
        if var>letr:
            features = [var, letr]
            x = df.loc[:, features].values
            y = df.loc[:,['Label']].values
            x = StandardScaler().fit_transform(x)
            from sklearn.decomposition import PCA
            pca = PCA(n_components=1)
            principalComponents = pca.fit_transform(x)
            principalDf = pd.DataFrame(data = principalComponents, columns = ['principal component 1'])
            finalDf = pd.concat([principalDf, df[['Label']]], axis = 1)
            p=pca.explained_variance_ratio_
            print('variance ratio on applying PCA on',var,'&',letr,'-',p)
```

```python
# On applying PCA to two features at a time and checking the variance ratio , we get the maximum variance ratio of 0.96538812 for feature 3 and feature 8.
# So, feature 3 and feature 8 are the best features which are same as those that we observed in Part 3.
```
