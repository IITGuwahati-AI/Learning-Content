```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
path = "https://raw.githubusercontent.com/eshantrehan26/Learning-Content/master/Phase%203%20-%202020%20(Summer)/Week%201%20(Mar%2028%20-%20Apr%204)/assignment/data.txt"

df = pd.read_csv(path,delim_whitespace =True)

lbl_1 = df[df["Label"]==1]
lbl_2 = df[df["Label"]==2]


for i in range(1,10):
    for j in range(i+1,11):



       plt.plot(lbl_1[str(i)],lbl_1[str(j)], 'o',c='red')
       plt.plot(lbl_2[str(i)],lbl_2[str(j)],'o',c='blue')
       plt.xlabel("Feature "+ str(i))
       plt.ylabel("Feature "+ str(j))
       plt.title("Feature "+str(i)+" vs "+"Feature "+str(j))

       plt.show()
```
