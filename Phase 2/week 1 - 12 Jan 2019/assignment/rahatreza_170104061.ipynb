```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
 a=np.loadtxt('data.txt',dtype=float,skiprows=1)  #loading dataset in numpy array
print(a)
```

```python
print(a.shape,a.ndim) #checkimng shape and dimension of array.
```

```python
k=a[:,0]==2     #making a boolean array named k
l=a[:,0]==1
for i in range(1,11):
    for j in range(1,11):
        if i!=j:
            
            plt.scatter(a[k,i],a[k,j],color='b',label='label2')   #a[k,i] is boolean array indexing
            plt.scatter(a[l,i],a[l,j],color='r',label='label1')
            plt.xlabel(f'feature{i}')
            plt.ylabel(f'feature{j}')
            plt.legend()
            plt.show()
```

```python
#claearly graph of feature1 vs feature2 is giving two labels classification perfectly.see again
plt.scatter(a[k,1],a[k,2],color='b',label='label2')
plt.scatter(a[l,1],a[l,2],color='r',label='label2')
```
