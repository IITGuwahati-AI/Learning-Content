```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
print("importing libraries successful")
```

```python
data = np.loadtxt(fname = 'data.txt', delimiter = '\t', skiprows = 1)
```

```python
df = pd.DataFrame(data)
```

```python
df_1 = df[df.ix[:,0]==1]
df_2 = df[df.ix[:,0]==2]
```

```python
for i in range(1,11):
    for j in range(i+1,11):
        plt.scatter(df_1.ix[:,i], df_1.ix[:,j],color = 'red')
        plt.scatter(df_2.ix[:,i], df_2.ix[:,j], color = 'blue')
        plt.xlabel("column "+str(i))
        plt.ylabel("column "+str(j))
        plt.title('Plot of columns '+str(i)+" vs "+str(j))
        plt.show()
        #plt.clf()
```
