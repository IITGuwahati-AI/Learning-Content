```python
import numpy as np
from matplotlib import pyplot as plt
data=np.loadtxt('ex1data1.txt',dtype='float',delimiter=",")
X,Y=data[:,0],data[:,1]
m,alpha=len(X),0.01
plt.scatter(X,Y,marker="x",color='r',label='Given Data')
plt.ylabel('Profit in $10,000')
plt.xlabel('Population of City in 10,000s')
r= np.linspace(5,22.5,10)
a,b,temp0,temp1=0,0,0,0
    
def Sum1(p,q,c,d,n):
    SUM=0
    for i in range(n):
        SUM=SUM+(c+(d*p[i])-q[i])
        
    return SUM    
    
def Sum2(p,q,c,d,n):
    SUM=0
    for i in range(n):
        SUM=SUM+(c+d*p[i]-q[i])*p[i]
    
    return SUM

for i in range(0,1500) :
    a=temp0
    b=temp1
    temp0=a-((alpha/m)*Sum1(X,Y,a,b,m))
    temp1=b-((alpha/m)*Sum2(X,Y,a,b,m))
   
    if temp1==b or temp0==a :
        break
        
plt.plot(r,a+(b*r),label='Linear Regression')
plt.legend()
plt.show()
print("The profit in a city of 35,000 is ${}".format((a+(3.5*b))*10000))
print("The profit in a city of 70,000 is ${}".format((a+(7*b))*10000))
```
