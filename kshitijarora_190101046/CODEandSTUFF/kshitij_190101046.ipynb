```python
import numpy as np
```

```python
from matplotlib import pyplot as plt
```

```python
data=np.loadtxt('data.txt', skiprows=1,delimiter='\t',dtype=np.float64)
```

```python
(num_rows,num_columns)=data.shape
label1=np.empty([1,11])
label2=np.empty([1,11])
for k in range(0,num_rows):
            if(data[k][0]==1):
                label1 = np.append(label1, np.array([data[k]]), axis=0)
            else:
                label2 = np.append(label2, np.array([data[k]]), axis=0)
count=0;
for i in range(1,10):
    for j in range(i+1,11):
        feature11=label1[:,i]
        feature21=label1[:,j]
        feature12=label2[:,i]
        feature22=label2[:,j]
        count+=1;
        plt.figure(count)
        plt.title('Label 1')
        plt.xlabel("feature number-"+str(i))
        plt.ylabel("feature number-"+str(j))
        plt.plot(feature11,feature21,'r')
        plt.savefig('1stLabel'+str(i)+'_'+str(j)+'.png')
        count+=1;
        plt.figure(count)
        plt.title('Label 2')
        plt.xlabel("feature number-"+str(i))
        plt.ylabel("feature number-"+str(j))
        plt.plot(feature12,feature22,'b')
        plt.savefig('2ndLabel'+str(i)+'_'+str(j)+'.png')
```
