```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib
```

```python
color=["red","blue"]
```

```python
y,data1,data2,data3,data4,data5,data6,data7,data8,data9,data10= np.loadtxt("data.txt",skiprows=1,unpack=True)
```

```python
data=[y,data1,data2,data3,data4,data5,data6,data7,data8,data9,data10]
```

```python
for i in range(9):
    plt.scatter(data[i+1],data[i+2],c=data[0],cmap=matplotlib.colors.ListedColormap(color))
    plt.xlabel('Feature1')
    plt.ylabel('Feature2')
    plt.title('feature1 vs feature2')
    plt.show()
```
